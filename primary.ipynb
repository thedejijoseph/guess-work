```python
import random

X = [i for i in range(1200)]
y = []

for i in X:
    random.seed(i)
    y.append(random.random())
```

```python
import sklearn
from sklearn import linear_model, utils

model = linear_model.LinearRegression()
dataset = utils.Bunch()

import numpy as np
dataset.data = np.array(X).reshape(-1, 1)
dataset.target = np.array(y).reshape(-1, 1)

model.fit(dataset.data, dataset.target)
# model.predict(13)

model.coef_
```

```python
from sklearn import metrics
import random

x_test = random.sample(range(1000), k=15)
y_test = []

for i in x_test:
    random.seed(i)
    y_test.append(random.random())

y_pred = model.predict(np.array(x_test).reshape(-1, 1))
cal_corl = metrics.r2_score
cal_corl(y_test, y_pred)
```
